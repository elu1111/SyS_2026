```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5.0, 5.0, 10000)
```

```python
fun = lambda v, lo, hi: 1 if lo < v <= hi else 0
modfun = lambda v, lo, hi: -((lo + hi)/2 + 5.0)*((lo + hi)/2 - 5.0)/4.0 if lo < v <= hi else 0
```

```python
plt.subplot(5,1,1)
plt.plot(x, [fun(v, -5.0, -3.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 1.5])
plt.subplot(5,1,2)
plt.plot(x, [fun(v, -3.0, -1.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 1.5])
plt.subplot(5,1,3)
plt.plot(x, [fun(v, -1.0, 1.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 1.5])
plt.subplot(5,1,4)
plt.plot(x, [fun(v, 1.0, 3.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 1.5])
plt.subplot(5,1,5)
plt.plot(x, [fun(v, 3.0, 5.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 1.5])
plt.savefig('pulses.png', format='png')
```

```python
plt.subplot(5,1,1)
plt.plot(x, [modfun(v, -5.0, -3.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 10.0])
plt.subplot(5,1,2)
plt.plot(x, [modfun(v, -3.0, -1.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 10.0])
plt.subplot(5,1,3)
plt.plot(x, [modfun(v, -1.0, 1.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 10.0])
plt.subplot(5,1,4)
plt.plot(x, [modfun(v, 1.0, 3.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 10.0])
plt.subplot(5,1,5)
plt.plot(x, [modfun(v, 3.0, 5.0) for v in x])
plt.axis([-5.0, 5.0, -0.5, 10.0])
plt.savefig('modulated_pulses.png', format='png')
```

```python
def graph(lo, hi):
    x1 = np.linspace(lo-0.1,hi+0.1,100)
    plt.plot(x1, [modfun(v, lo, hi) for v in x1])    
```

```python
plt.clf()
graph(-5.0,-3.0)
graph(-3.0,-1.0)
graph(-1.0,1.0)
graph(1.0,3.0)
graph(3.0,5.0)
plt.plot(x, [-(v + 5.0)*(v - 5.0)/4.0 for v in x], 'k')
plt.savefig('coarse-approximation.png', format='png')
```

```python
plt.clf()
points = np.linspace(-5.0, 5.0, 50)
for lo, hi in zip(points[:-1], points[1:]):
  graph(lo, hi)
plt.savefig('fine-approximation.png', format='png')
```
